```python
import pandas as pd 
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(low=20, high=36, size=(8, 1)), columns=['age'])
df['gender'] = np.random.choice(['male', 'female'], size=(8))
df['salary'] = np.random.randint(low=2000, high=8001, size=(8))
df['job_class'] = np.random.choice(['white-collar', 'blue-collar' , "meneger" , "CEO"], size=(8))
df['position'] = np.random.choice(['manager', 'supervisor', 'assistant', 'clerk'], size=(8,))
df['is_married'] = np.random.choice([True, False], size=(8,))
df.index = ['ali', 'mehmet', 'metehan', 'berfin', 'emine', 'emre', 'şükür', 'hüseyin']
df
```
